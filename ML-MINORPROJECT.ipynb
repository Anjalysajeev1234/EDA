## Q3) Most Common Genre for Emma.W and V.Diesel

```python
query_data = df.filter(['cast','genres'])
Vin_Diesel_movies = query_data[query_data.cast.str.contains('Vin Diesel',na=False)]
Emma_Watson_movies = query_data[query_data.cast.str.contains('Emma Watson',na=False)]
```

```python
print("VinDiesel : ")
print("{}".format(Vin_Diesel_movies.count()))
print("EmmaWatson :")
print("{}".format(Emma_Watson_movies.count()))
```

```python
def listify(genre):
  return genre.split('|')
EW_genres = Emma_Watson_movies.genres.apply(listify)
VD_genres = Vin_Diesel_movies.genres.apply(listify)
```

```python
EW_genres
```

```python
EW_genre_count = dict()
for genres in EW_genres:
  for genre in genres:
    if genre in EW_genre_count.keys():
      EW_genre_count[genre] +=1
    else:
      EW_genre_count[genre] =1

VD_genre_count = dict()
for genres in VD_genres:
  for genre in genres:
    if genre in VD_genre_count.keys():
      VD_genre_count[genre] +=1
    else:
      VD_genre_count[genre] =1
```

```python
common_genres_sorted_EW = sorted(EW_genre_count.items(), key=lambda item: item[1], reverse=True)
common_genres_sorted_VD = sorted(VD_genre_count.items(), key=lambda item: item[1], reverse=True)
```

```python
common_genres_sorted_EW
```

```python
common_genres_sorted_VD
```

```python
print("The most common genre for the movies starring Emma Watson is {}".format(common_genres_sorted_EW[0][0]))
print("The most common genre for the movies starring Vin Diesel is {}".format(common_genres_sorted_VD[0][0]))
```
